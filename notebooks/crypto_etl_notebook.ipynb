```python
from pyspark.sql.functions import *

# Read Bronze Layer
df_bronze = spark.read.json(
    "abfss://crypto-data@stcryptoeng001.dfs.core.windows.net/bronze/crypto_market.json"
)

# Create Silver Layer
df_silver = df_bronze.select(
    col("id"),
    upper(col("symbol")).alias("symbol"),
    col("name"),
    col("current_price").cast("double"),
    col("market_cap").cast("long"),
    col("market_cap_rank").cast("int"),
    col("total_volume").cast("long"),
    col("price_change_percentage_24h").cast("double"),
    col("high_24h").cast("double"),
    col("low_24h").cast("double"),
    col("ath").cast("double"),
    col("atl").cast("double"),
    col("circulating_supply").cast("double"),
    current_timestamp().alias("ingestion_time")
)

# Add Market Movement Category
df_silver = df_silver.withColumn(
    "market_status",
    when(col("price_change_percentage_24h") > 5, "Strong Bullish")
    .when(col("price_change_percentage_24h") > 0, "Bullish")
    .when(col("price_change_percentage_24h") < -5, "Strong Bearish")
    .otherwise("Bearish")
)

# Add Price Category
df_silver = df_silver.withColumn(
    "price_category",
    when(col("current_price") >= 1000, "Very High")
    .when(col("current_price") >= 100, "High")
    .when(col("current_price") >= 1, "Medium")
    .otherwise("Low")
)

# Add Volume Ratio KPI
df_silver = df_silver.withColumn(
    "volume_marketcap_ratio",
    round(col("total_volume") / col("market_cap"), 4)
)

# Show transformed data
display(df_silver)
```

```python
from pyspark.sql.functions import current_date

# Add processing date column
df_silver = df_silver.withColumn(
    "processing_date",
    current_date()
)

# Append transformed data to Silver Layer
df_silver.write \
    .mode("append") \
    .partitionBy("processing_date") \
    .parquet(
        "abfss://crypto-data@stcryptoeng001.dfs.core.windows.net/silver/crypto_market_clean"
    )
```

```python
from pyspark.sql.functions import *

# =====================================================
# READ SILVER LAYER
# =====================================================

df_gold_source = spark.read.parquet(
    "abfss://crypto-data@stcryptoeng001.dfs.core.windows.net/silver/crypto_market_clean"
)

# =====================================================
# ADD PROCESSING COLUMNS
# =====================================================

df_gold_source = df_gold_source.withColumn(
    "processed_date",
    current_date()
)

df_gold_source = df_gold_source.withColumn(
    "processed_time",
    current_timestamp()
)

# =====================================================
# CREATE GOLD KPI TABLES
# =====================================================

# TOP GAINERS
top_gainers = df_gold_source.orderBy(
    col("price_change_percentage_24h").desc()
).limit(10)

# TOP LOSERS
top_losers = df_gold_source.orderBy(
    col("price_change_percentage_24h").asc()
).limit(10)

# TOP MARKET CAP
top_marketcap = df_gold_source.orderBy(
    col("market_cap").desc()
).limit(10)

# TOP VOLUME
top_volume = df_gold_source.orderBy(
    col("total_volume").desc()
).limit(10)

# MARKET SUMMARY
market_summary = df_gold_source.groupBy(
    "market_status"
).agg(
    count("*").alias("total_coins"),
    round(avg("price_change_percentage_24h"), 2).alias("avg_change_pct"),
    round(avg("current_price"), 2).alias("avg_price")
)

# =====================================================
# IMPORTANT FIX FOR MARKET SUMMARY
# =====================================================

market_summary = market_summary.withColumn(
    "processed_date",
    current_date()
)

# =====================================================
# DISPLAY KPI TABLES
# =====================================================

print("TOP GAINERS")
display(top_gainers)

print("TOP LOSERS")
display(top_losers)

print("TOP MARKET CAP")
display(top_marketcap)

print("TOP VOLUME")
display(top_volume)

print("MARKET SUMMARY")
display(market_summary)

# =====================================================
# SAVE LATEST SNAPSHOT TABLES
# =====================================================

top_gainers.write \
    .mode("overwrite") \
    .parquet(
        "abfss://crypto-data@stcryptoeng001.dfs.core.windows.net/gold/latest/top_gainers"
    )

top_losers.write \
    .mode("overwrite") \
    .parquet(
        "abfss://crypto-data@stcryptoeng001.dfs.core.windows.net/gold/latest/top_losers"
    )

top_marketcap.write \
    .mode("overwrite") \
    .parquet(
        "abfss://crypto-data@stcryptoeng001.dfs.core.windows.net/gold/latest/top_marketcap"
    )

top_volume.write \
    .mode("overwrite") \
    .parquet(
        "abfss://crypto-data@stcryptoeng001.dfs.core.windows.net/gold/latest/top_volume"
    )

market_summary.write \
    .mode("overwrite") \
    .parquet(
        "abfss://crypto-data@stcryptoeng001.dfs.core.windows.net/gold/latest/market_summary"
    )

# =====================================================
# SAVE HISTORICAL TRACKING TABLES
# =====================================================

top_gainers.write \
    .mode("append") \
    .partitionBy("processed_date") \
    .parquet(
        "abfss://crypto-data@stcryptoeng001.dfs.core.windows.net/gold/historical/top_gainers"
    )

top_losers.write \
    .mode("append") \
    .partitionBy("processed_date") \
    .parquet(
        "abfss://crypto-data@stcryptoeng001.dfs.core.windows.net/gold/historical/top_losers"
    )

top_marketcap.write \
    .mode("append") \
    .partitionBy("processed_date") \
    .parquet(
        "abfss://crypto-data@stcryptoeng001.dfs.core.windows.net/gold/historical/top_marketcap"
    )

top_volume.write \
    .mode("append") \
    .partitionBy("processed_date") \
    .parquet(
        "abfss://crypto-data@stcryptoeng001.dfs.core.windows.net/gold/historical/top_volume"
    )

market_summary.write \
    .mode("append") \
    .partitionBy("processed_date") \
    .parquet(
        "abfss://crypto-data@stcryptoeng001.dfs.core.windows.net/gold/historical/market_summary"
    )

print("Gold Layer Saved Successfully")
```
